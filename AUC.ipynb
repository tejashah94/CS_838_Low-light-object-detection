```python
def bb_intersection_over_union(boxA, boxB):
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
 
    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
 
    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
 
    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)
 
    # return the intersection over union value
    return iou
```

```python
def acquire_gt(img_pred):
    
    # img_path operation
    path = '.'+img_pred['path'][18:]
    path = path.replace('basic_enhance_', '')
    path = path.replace('ExDark', 'ExDark_Anno')+'.txt'
    path = path.replace('objecti_detection/', '')

    try:
        gt = np.loadtxt(path, dtype=np.ndarray, skiprows=1)
    except:
        try:
            gt = np.loadtxt(path.replace("jpg", "png"), dtype=np.ndarray, skiprows=1)
        except:
            return None, None
    
    gt_center_hub = []
    gt_class_hub = []
    gt_box_hub = []
    if len(gt.shape) == 1:
        gt = [gt]
    
    try:
        for obj_gt in gt:
            gt_class = obj_gt[0]
    except:
        return None, None
    
    for obj_gt in gt:
        gt_class = obj_gt[0]
        gt_class_hub.append(gt_class)
        l, t, w, h = int(obj_gt[1]), int(obj_gt[2]), int(obj_gt[3]), int(obj_gt[4])

        gt_center = [l+w/2, t+h/2]
        gt_center_hub.append(gt_center)

        ymin, xmin, ymax, xmax = t, l, t+h, l+w
        gt_box_hub.append([ymin, xmin, ymax, xmax])

    return gt_class_hub, gt_box_hub
```

```python
def acquire_pred(img_pred, height, width):
    pred = img_pred['result']
    pred_center_hub = []
    pred_class_hub = []
    pred_box_hub = []
    for idx, pred_score in enumerate(pred["detection_scores"]):
        if pred_score > 0.5:
            pred_class = coco_index[pred["detection_classes"][idx]]["name"]
            pred_class_hub.append(pred_class)

            # ymin, xmin, ymax, xmax
            ymin, xmin, ymax, xmax = pred["detection_boxes"][idx]
            ymin, ymax = int(ymin*height), int(ymax*height)
            xmin, xmax = int(xmin*width), int(xmax*width)
            pred_center = [(xmin+xmax)/2, (ymin+ymax)/2]
            pred_center_hub.append(pred_center)

            pred_box_hub.append([ymin, xmin, ymax, xmax])
    
    return pred_class_hub, pred_box_hub
```

```python
def acquire_dim_info(img_pred):
    # load dimension info
    filename = os.path.basename(img_pred['path'])[:-4]
    if filename in dim_hub.keys():
        dim_info = dim_hub[filename]
    else:
        return None, None
    height, width = dim_info[0], dim_info[1]
    return height, width
```

```python
def acquire_recall_precision(pred_list):
    result_hub = {}
    for img_pred in pred_list:
        filename = os.path.basename(img_pred['path'])[:-4]
        # dim info
        height, width = acquire_dim_info(img_pred)
        if height is None:
            continue

        # ground truth
        gt_class_hub, gt_box_hub = acquire_gt(img_pred)
        if gt_class_hub is None:
            continue

        # prediction
        pred_class_hub, pred_box_hub = acquire_pred(img_pred, height, width)

        # Calculate precision and recall
        iter_result = dict()
        for gt_idx, gt_class in enumerate(gt_class_hub):
            if not gt_class in iter_result:
                iter_result[gt_class] = {'TP':0, "FP":0, 'FN':0}
            gt_box = gt_box_hub[gt_idx]


            # in each obj in gt, we are finding the best estimation
            # we calculate IOU for all predtions, and take the maximum

            IOU_hub = [0]
            for pred_idx, pred_class in enumerate(pred_class_hub):
                IOU = bb_intersection_over_union(gt_box, pred_box_hub[pred_idx])
                if pred_class in ExDark_tag_hub[gt_class] and IOU > 0.5:
                    IOU_hub.append(IOU)
                else:
                    IOU_hub.append(0)
            IOU_hub = np.asarray(IOU_hub)
            if np.amax(IOU_hub)>0.5:
                loc_max_IOU = np.argmax(IOU_hub)
                value_max_IOU = IOU_hub[loc_max_IOU]

                # if maximum > 0.5, identify as TP and remove it from pred_
                # if maximum < 0.5, identify as FN
                # when finish all gt, the rest pred_ is FP
                # so recall = TP / len(gt)=(TP+FN), class-wise
                # and precision = TP / len(pred)=(TP+FP), class-wisc
                iter_result[gt_class]['TP'] += 1
                del pred_class_hub[pred_idx]
                del pred_box_hub[pred_idx]
            else:
                iter_result[gt_class]['FN'] += 1

        # calculate FP
        unique_gt_class_hub = np.unique(np.array(gt_class_hub))
        for unique_gt_class in unique_gt_class_hub:
            for pred_class in pred_class_hub:
                if pred_class in ExDark_tag_hub[unique_gt_class]:
                    iter_result[unique_gt_class]['FP'] += 1

        # calculate recall/precision in this image:
        for unique_gt_class in unique_gt_class_hub:
            if not unique_gt_class in result_hub:
                    result_hub[unique_gt_class] = []
            if iter_result[unique_gt_class]['TP'] == 0:
                result_hub[unique_gt_class].append({"recall":0, "precision":0})
            else:
                TP = iter_result[unique_gt_class]['TP']
                FN = iter_result[unique_gt_class]['FN']
                FP = iter_result[unique_gt_class]['FP']
                recall = TP/(TP+FN)
                precision = TP/(TP+FP)
                result_hub[unique_gt_class].append({"recall":recall, 
                                                     "precision":precision, 
                                                     "filename":filename})
    return result_hub
```

```python
def mAP(result_hub):
    AP = result_hub
    AP_hub = []
    for ins_AP in AP:
        recall = []
        precision = []
        curve = []

        for ins_img in AP[ins_AP]:
            recall.append(ins_img["recall"])
            precision.append(ins_img["precision"])

        recall = np.array(recall)
        precision = np.array(precision)

        recall_hub = np.unique(recall)
        for atom_recall in recall_hub:
            # find max precision over atom_recall
            temp_precision = copy.deepcopy(precision)
            temp_precision = precision[recall==atom_recall]
            max_precision = np.max(temp_precision)
            curve.append([atom_recall, max_precision])
        curve = np.array(curve)
        classAP = np.mean(curve[:, 0]*curve[:, 1])
#         print(ins_AP, classAP)
        AP_hub.append(classAP)
    mAP = np.mean(np.asarray(AP_hub))
#     print("mAP: ", mAP)
    return mAP
```
